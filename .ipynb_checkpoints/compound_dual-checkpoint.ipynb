# Optimal design for heteroscedastic polynomial models:

Given a compact design space $\mathcal{X}$ that have the form 
$$
    y(x)=g^\top(x) \beta+ \frac{e(x)}{\sqrt{\lambda(x)}}
$$


$\newcommand{\Z}{\mathcal Z}$
$\newcommand{\X}{\mathcal X}$
$\newcommand{\bw}{\mathbf w}$
$\newcommand{\tr}{\operatorname{tr}}$
$\newcommand{\logdet}{\log~\det}$
$\newcommand{\inv}{\operatorname{inv}}$

## In this notebook, we consider the compound optimal design.

The objective function is:
$$
    \sum_{k=1}^K \alpha_k \phi_k(M),\quad \sum \alpha_k = 1,
$$
where $M$ is the Fisher information matrix.

In the example below, we consider compound dual-optimality for $D$- and $A$-optimality.


### 1. Dual optimality for $D$-$A$-optimality.

The optimization problem is 
$$
\begin{aligned}
    \min_{\boldsymbol{w}} \quad & -\alpha \cdot \log\det(M) + (1-\alpha)\operatorname{tr}(M^{-1}) \\
    \text{subject to} \quad & \sum_i w_i = 1, \\
                            & w_i \ge 0,\quad \forall i,
\end{aligned}
$$
where $\alpha\in[0,1]$

```python
import numpy as np
import matplotlib.pyplot as plt
import cvxpy as cp

# === Settings ===
g = lambda x: np.array([1, x, x**2])
lambda_func = lambda x: 2 * x + 5  # Heteroscedastic
N = 501
x_vals = np.linspace(-1, 1, N)
q = 3
alpha = 0.5  # Weight for A-optimality
```

```python
# === Step 1: Precompute g(x) and lambda(x) ===
g_list = [g(x) for x in x_vals]
lambda_vals = np.array([lambda_func(x) for x in x_vals])

# === Step 2: Solve compound criterion via CVXPY ===
w = cp.Variable(N)
M = sum(w[i] * lambda_vals[i] * np.outer(g_list[i], g_list[i]) for i in range(N))

# Use matrix_frac(I, M) to compute trace(inv(M))
I = np.eye(q)
trace_inv_expr = cp.matrix_frac(I, M)

# Compound objective
objective = alpha * trace_inv_expr - (1 - alpha) * cp.log_det(M)
constraints = [cp.sum(w) == 1, w >= 0]
prob = cp.Problem(cp.Minimize(objective), constraints)
prob.solve()

# === Extract weights and support points ===
w_opt = w.value
support_idx = np.where(w_opt > 1e-4)[0]
x_out = np.round(x_vals[support_idx], 3)
w_out = np.round(w_opt[support_idx], 3)

print("Support points and weights:")
for xi, wi in zip(x_out, w_out):
    print(f"x = {xi:.3f}, weight = {wi:.3f}")
```

```python
# === Step 3: Compute Fisher info and inverse ===
M_val = sum(w_opt[i] * lambda_vals[i] * np.outer(g_list[i], g_list[i]) for i in range(N))
M_inv = np.linalg.inv(M_val)
M_inv2 = M_inv @ M_inv

# === Step 4: Directional Derivatives ===
phiD = np.zeros(N)
phiA = np.zeros(N)
phiMix = np.zeros(N)

for i in range(N):
    gx = g_list[i]
    phiD[i] = lambda_vals[i] * (gx @ M_inv @ gx) - q
    phiA[i] = lambda_vals[i] * (gx @ M_inv2 @ gx) - np.trace(M_inv)
    phiMix[i] = alpha * phiA[i] + (1 - alpha) * phiD[i]
```

### Plotting the equivalence theorem

```python
# === Step 5: Plot 1x3 subplots ===
fontsize = 24
fig, axs = plt.subplots(1, 3, figsize=(20, 6))

# --- Subplot 1: D-optimality ---
axs[0].plot(x_vals, phiD, 'k-', linewidth=2)
axs[0].plot(x_vals[support_idx], phiD[support_idx], 'ro', markersize=6)
axs[0].axhline(0, color='g', linewidth=2)
axs[0].set_xlabel('x', fontsize=fontsize)
axs[0].set_ylabel(r'$\phi_D(x)$', fontsize=fontsize)
axs[0].set_title('D-opt Directional Derivative', fontsize=fontsize, fontweight='bold')
axs[0].grid(True)
axs[0].tick_params(labelsize=fontsize)

# --- Subplot 2: A-optimality ---
axs[1].plot(x_vals, phiA, 'k-', linewidth=2)
axs[1].plot(x_vals[support_idx], phiA[support_idx], 'ro', markersize=6)
axs[1].axhline(0, color='g', linewidth=2)
axs[1].set_xlabel('x', fontsize=fontsize)
axs[1].set_ylabel(r'$\phi_A(x)$', fontsize=fontsize)
axs[1].set_title('A-opt Directional Derivative', fontsize=fontsize, fontweight='bold')
axs[1].grid(True)
axs[1].tick_params(labelsize=fontsize)

# --- Subplot 3: Mix (A+D) ---
axs[2].plot(x_vals, phiMix, 'k-', linewidth=2)
axs[2].plot(x_vals[support_idx], phiMix[support_idx], 'ro', markersize=6)
axs[2].axhline(0, color='g', linewidth=2)
axs[2].set_xlabel('x', fontsize=fontsize)
axs[2].set_ylabel(r'$\phi_{\mathrm{mix}}(x), \Delta(x)$', fontsize=fontsize)
axs[2].set_title(f'Weighted A+D (α = {alpha:.2f})', fontsize=fontsize, fontweight='bold')
axs[2].legend([r'$\phi_{\mathrm{mix}}(x)$', r'$\Delta(x)$'], loc='best', fontsize=fontsize)
axs[2].grid(True)
axs[2].tick_params(labelsize=fontsize)

plt.tight_layout()
plt.show()
```

## 2. D-E optimality

```python
N = 501
x_vals = np.linspace(-1, 1, N)
q = 3
alpha = 0.5

# === Step 1: Precompute g(x) and lambda(x) ===
g_list = [g(x) for x in x_vals]
lambda_vals = np.array([lambda_func(x) for x in x_vals])

# === Step 2: Solve weighted E- and D-optimal design via CVXPY ===
w = cp.Variable(N)
M = sum(w[i] * lambda_vals[i] * np.outer(g_list[i], g_list[i]) for i in range(N))

# Objective: -alpha * lambda_min(M) - (1 - alpha) * log_det(M)
objective = -alpha * cp.lambda_min(M) - (1 - alpha) * cp.log_det(M)
constraints = [cp.sum(w) == 1, w >= 0]
prob = cp.Problem(cp.Minimize(objective), constraints)
prob.solve(solver=cp.SCS, verbose=False)

w_opt = w.value
support_idx = np.where(w_opt > 1e-4)[0]
x_out = np.round(x_vals[support_idx], 3)
w_out = np.round(w_opt[support_idx], 3)
print("Support points and weights:")
for x, w in zip(x_out, w_out):
    print(f"({float(x):.3f}, {float(w):.3f})")

# === Step 3: Compute Fisher information ===
M_val = sum(w_opt[i] * lambda_vals[i] * np.outer(g_list[i], g_list[i]) for i in range(N))
M_inv = np.linalg.inv(M_val)
M_inv2 = M_inv @ M_inv
```

```python
# === Step 4: Directional Derivatives ===
phiD = np.zeros(N)
phiE = np.zeros(N)
phiMix = np.zeros(N)

eigvals, eigvecs = np.linalg.eigh(M_val)
lambda_min_E = np.min(eigvals)
tol = 1e-4
Q = eigvecs[:, np.abs(eigvals - lambda_min_E) < tol]  # eigenvectors for min eig

for i in range(N):
    gx = g_list[i]
    lam = lambda_vals[i]
    phiD[i] = lam * gx.T @ M_inv @ gx - q
    phiE[i] = lam * gx.T @ Q @ Q.T @ gx - lambda_min_E
    phiMix[i] = alpha * phiE[i] + (1 - alpha) * phiD[i]

# === Step 5: Plot ===
fontsize = 24
fig, axs = plt.subplots(1, 3, figsize=(20, 6))

# Subplot 1: D-opt
axs[0].plot(x_vals, phiD, 'k-', linewidth=1.5)
axs[0].plot(x_vals[support_idx], phiD[support_idx], 'ro', markersize=6)
axs[0].axhline(0, color='g', linewidth=3)
axs[0].set_xlabel('x', fontsize=fontsize)
axs[0].set_ylabel(r'$\phi_D(x)$', fontsize=fontsize)
axs[0].set_title('D-opt Directional Derivative', fontsize=fontsize, fontweight='bold')
axs[0].grid(True)
axs[0].tick_params(labelsize=fontsize)

# Subplot 2: E-opt
axs[1].plot(x_vals, phiE, 'k-', linewidth=1.5)
axs[1].plot(x_vals[support_idx], phiE[support_idx], 'ro', markersize=6)
axs[1].axhline(0, color='g', linewidth=3)
axs[1].set_xlabel('x', fontsize=fontsize)
axs[1].set_ylabel(r'$\phi_E(x)$', fontsize=fontsize)
axs[1].set_title('E-opt Directional Derivative', fontsize=fontsize, fontweight='bold')
axs[1].grid(True)
axs[1].tick_params(labelsize=fontsize)

# Subplot 3: Weighted A+E
axs[2].plot(x_vals, phiMix, 'k-', linewidth=2)
axs[2].plot(x_vals[support_idx], phiMix[support_idx], 'ro', markersize=6)
axs[2].axhline(0, color='g', linewidth=3)
axs[2].set_xlabel('x', fontsize=fontsize)
axs[2].set_ylabel(r'$\phi_{\mathrm{mix}}(x), \Delta(x)$', fontsize=fontsize)
axs[2].set_title(f'Weighted D+E ($\\alpha = {alpha:.2f}$)', fontsize=fontsize, fontweight='bold')
axs[2].legend([r'$\phi_{\mathrm{mix}}(x)$', r'$\Delta(x)$'], loc='best', fontsize=fontsize)
axs[2].grid(True)
axs[2].tick_params(labelsize=fontsize)

plt.tight_layout()
plt.show()
```
