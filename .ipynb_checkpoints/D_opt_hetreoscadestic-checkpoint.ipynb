```python
import numpy as np
import cvxpy as cp
import matplotlib.pyplot as plt

# === Settings ===
def g(x):
    return np.array([1, x, x**2])

# Choose different forms of the lambda function
def lambda_uniform(x):
    return 1

def lambda_linear(x):
    return 2 * x + 5

# Switch the lambda function to use
lambda_func = lambda_linear  # or lambda_uniform

N = 101
x_vals = np.linspace(-1, 1, N)
n = len(x_vals)
q = 3

g_list = [g(x) for x in x_vals]
lambda_vals = np.array([lambda_func(x) for x in x_vals])

# === Solve D-optimal design using cvxpy ===
w = cp.Variable(n)
# === Vectorized version ===
G = np.stack(g_list)  # G.shape = (n, q)
W = cp.multiply(w, lambda_vals)  # Elementwise weighting

# Construct M as G.T @ diag(W) @ G
M = G.T @ cp.diag(W) @ G
objective = cp.Maximize(cp.log_det(M))
constraints = [w >= 0, cp.sum(w) == 1]
prob = cp.Problem(objective, constraints)

# Try using the MOSEK solver
# try:
#     prob.solve(solver=cp.MOSEK, mosek_params={
#         "MSK_DPAR_INTPNT_CO_TOL_REL_GAP": 1e-6,
#         "MSK_DPAR_INTPNT_CO_TOL_PFEAS": 1e-6,
#         "MSK_DPAR_INTPNT_CO_TOL_DFEAS": 1e-6,
#         "MSK_DPAR_INTPNT_CO_TOL_MU_RED": 1e-6,
#         "MSK_DPAR_INTPNT_TOL_INFEAS": 1e-6},
#         verbose=True)
#         # "MSK_IPAR_INTPNT_MAX_ITERATIONS": 1000
# except (cp.SolverError, ImportError) as e:
#     if w.value is not None:
#         print("MOSEK technically succeeded despite warning.")
#     else:
#         print("MOSEK failed. Reason:", e)
#         prob.solve()
prob.solve()

# === Extract the design weights ===
w_val = w.value

# === Identify support points and weights ===
support_idx = np.where(w_val > 1e-3)[0]
support_x = x_vals[support_idx]
support_w = w_val[support_idx]/sum(w_val[support_idx])

# === Display support points and weights ===
print("D-optimal Design Support Points and Weights:")
for xi, wi in zip(support_x, support_w):
    print(f"x = {xi:.3f}, weight = {wi:.3f}")
```

# Dispersion function

```python
# === 計算 d_D(x) ===
M_val = sum(w_val[i] * lambda_vals[i] * np.outer(g_list[i], g_list[i]) for i in range(n))
Minv = np.linalg.inv(M_val)

dD = np.array([g_list[i].T @ Minv @ g_list[i] * lambda_vals[i] - q for i in range(n)])
```

# Plotting

```python
# === 畫圖 ===
support_idx = np.where(w_val > 1e-4)[0]
support_x = x_vals[support_idx]
support_w = w_val[support_idx]
design = np.vstack([support_x, support_w])

plt.figure(figsize=(10, 6))
plt.plot(x_vals, dD, 'b-', linewidth=2, label='d_D(x)')
plt.axhline(0, color='k', linestyle='--', linewidth=1.5, label='Zero line')
for xj in support_x:
    plt.axvline(xj, color='r', linestyle='--', linewidth=1.2)
plt.xlabel('x')
plt.ylabel('d_D(x)')
plt.title('D-optimality Equivalence Function with Heteroscedastic Errors')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
